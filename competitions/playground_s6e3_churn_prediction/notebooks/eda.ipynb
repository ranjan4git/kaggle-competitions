```python
import pandas as pd
import numpy as np

import joblib
```

```python
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')
```

```python
train_df[target_col] = train_df[target_col].map({"Yes": 1, "No": 0})
train_df[target_col].value_counts()
```

##### Creating a baseline Random Forest model strictly using numerical data

train df
keep numericals only
remove id 
remove churn from X, keep in y
train random forest


```python
num_cols = train_df.select_dtypes(include=np.number).columns.to_list()
num_cols
```

```python
X_train = train_df[[col for col in num_cols if col not in ['id', 'Churn']]]
y_train = train_df[target_col]
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
joblib.dump(rf, "../models/baseline_rf_num_only.joblib", compress=3)
```

```python
rf = joblib.load("../models/baseline_rf_num_only.joblib")
```

```python
test_df
```

```python
X_test = test_df[[col for col in num_cols if col not in ['id', 'Churn']]]
X_test.head()
```

```python
y_proba = rf.predict_proba(X_test)
y_proba
```

```python
submission = test_df.copy()
submission['Churn'] = y_proba[:,1]
submission = submission[['id', 'Churn']]
submission.head()
```

I see multiple categorical columns with Yes/No values. I can either use a categorical-friendly model, or I can convert the yes/no to numeric binary using
`df[<categorical_column>] = df[<categorical_column>].apply(lambda x: 1 if x=="Yes" else 0)` *(slower)* <br>
or <br>
`df.loc[df[<categorical_column>] == "Yes", <categorical_column>] = 1`

`df.loc[df[<categorical_column>] == "No", <categorical_column>] = 0` *(vectorized, faster)*
<br> or <br>
`df[<categorical_column>] = (df[<categorical_column>] == "Yes").astype(int) `
